# Gasoline Price Predictions 

## Pre-processing Preparing

## Average Price Prediction

* Checking if `gasolina_comum_preco_revenda_avg` have null values

```python
df.gasolina_comum_preco_revenda_avg.isnull().sum()
```

* Creating a new dataframe to analysis segment using only index and metrics value

```python
df_gasolina_avg = df.gasolina_comum_preco_revenda_avg
```

```python
df_gasolina_avg.tail()
```

### Creating Train and Test dataframes

```python
train = df_gasolina_avg[df_gasolina_avg.index <= pd.to_datetime("2022-09-01", format='%Y-%m-%d')]
train.index = pd.DatetimeIndex(train.index.values, freq=train.index.inferred_freq)
```

```python
train.index = train.index.to_period('M')
```

```python
train.shape
```

```python
test = df_gasolina_avg[df_gasolina_avg.index >= pd.to_datetime("2022-09-01", format='%Y-%m-%d')]
test.index = pd.DatetimeIndex(test.index.values,
                               freq=test.index.inferred_freq)
```

```python
test.index = test.index.to_period('M')
```

```python
test.shape
```

### Creating Training and Test Visualization

```python
x_train = pd.to_datetime(train.index.to_timestamp())
y_train = train.values
```

```python
x_test = pd.to_datetime(test.index.to_timestamp())
y_test = test.values
```

```python
plt.figure(figsize=(25, 8))
plt.plot(x_train, y_train, label="Gasoline - Train", color='black')
plt.plot(x_test, y_test, label="Gasoline - Test", color='red')
plt.xlabel("Reference Period")
plt.ylabel("Gasoline Price")


plt.title("Gasoline Price Train and Test")
plt.legend()
plt.show()
```

### Finding best fit to ARIMA model
* The Gas Prices is not Stationary, in this case we need to set `seasonal=False` to find the best fit model
* I used `autoarima` to find the best fit
* Kaggle don't have pmarima available, but the function returned Best ARIMA Model `(2,1,3)` 
* [I published my jupyter notebook with complete execution]()

### Testing SARIMAX models 

```python
SARIMAXmodel = SARIMAX(train, order=(2,1,3), seasonal_order=(0,0,0,0), enforce_stationarity=False, enforce_invertibility=False)
SARIMAXmodel = SARIMAXmodel.fit()
```

```python
y_pred = SARIMAXmodel.get_forecast(len(test.index))
y_pred_df = y_pred.conf_int(alpha = 0.05) 
y_pred_df["Predictions"] = SARIMAXmodel.predict(start = y_pred_df.index[0], end = y_pred_df.index[-1])
y_pred_df.index = test.index

y_pred_out_sarima = y_pred_df["Predictions"] 
```

```python
y_pred_out_sarima
```

```python
plt.figure(figsize=(25, 8))

plt.plot(x_train, y_train, label="Gasoline - Train", color='black')
plt.plot(x_test, y_test, label="Gasoline - Test", color='red')
plt.plot(y_pred_out_sarima, color='Blue', label = 'Gas Predictions SARIMA to Test')

plt.xlabel("Reference Period")
plt.ylabel("Gasoline Price")

plt.title("Gasoline Price Train and Test")
plt.legend()
plt.show()
```

### Create Predictions to 12 Months Forecast

```python
start_predict = pd.DatetimeIndex(["2022-09-01"], freq="MS")[0]
start_predict
```

```python
end_predict = pd.DatetimeIndex(["2023-12-01"], freq="MS")[0]
end_predict
```

```python
prediction_sarimax = SARIMAXmodel.predict(start=start_predict, end=end_predict)
```
